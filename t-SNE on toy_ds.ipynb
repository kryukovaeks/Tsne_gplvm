```python
import os

os.environ['KMP_DUPLICATE_LIB_OK']='True'
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
%matplotlib inline

sizeTitle = 24
sizeAxis = 14
```

# #Cylinder dataset

```python
#Loading cylinder datasets
cylinder = np.load("cylinder.npy")
cylinder_full = np.load("cylinder_full.npy")

fig = plt.figure(figsize = plt.figaspect(0.5))
matplotlib.rcParams.update({'font.size':sizeAxis})
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.scatter(cylinder_full[:,0], cylinder_full[:,1], cylinder_full[:,2], c=cylinder_full[:,3])
plt.title('Full', fontsize=sizeTitle)
ax = fig.add_subplot(1, 2, 2,  projection='3d')
ax.scatter(cylinder[:,0], cylinder[:,1], cylinder[:,2], c=cylinder[:,3])
plt.title('Slit', fontsize=sizeTitle)
plt.show()
```

```python
#TSNE for various perplexity and iterations' number (cylinder with slit)
for perp in range(5, 50, 5):
    for iter in range(250, 650, 50):
        tsne = TSNE(n_components=2, verbose=1, perplexity=perp, n_iter=iter)
        tsne_results = tsne.fit_transform(cylinder[:,0:3])
        #plots
        fig = plt.figure(figsize = plt.figaspect(0.5))
        matplotlib.rcParams.update({'font.size':sizeAxis})
        ax = fig.add_subplot(1, 2, 1)
        ax.scatter(tsne_results[:,0], tsne_results[:,1], c=cylinder[:,3])
        plt.title('TSNE: perp = %d' %(perp)+', iter = %d' %(iter), fontsize=sizeTitle)
        plt.show()
```

```python
#TSNE for a single set of hyper-parameters (cylinder with slit)
tsne = TSNE(n_components=2, verbose=1, perplexity=50, n_iter=800)
tsne_results = tsne.fit_transform(cylinder[:,0:3])

fig = plt.figure(figsize = plt.figaspect(0.5))
matplotlib.rcParams.update({'font.size':sizeAxis})
ax = fig.add_subplot(1, 2, 1)
ax.scatter(tsne_results[:,0], tsne_results[:,1], c=cylinder[:,3])
plt.title('TSNE', fontsize=sizeTitle)

plt.show()
```

```python
#TSNE for various perplexity and iterations' number (full cylinder)
for perp in range(5, 50, 5):
    for iter in range(250, 650, 50):
        tsne = TSNE(n_components=2, verbose=1, perplexity=perp, n_iter=iter)
        tsne_results = tsne.fit_transform(cylinder_full[:,0:3])
        #plots
        fig = plt.figure(figsize = plt.figaspect(0.5))
        matplotlib.rcParams.update({'font.size':sizeAxis})
        ax = fig.add_subplot(1, 2, 1)
        ax.scatter(tsne_results[:,0], tsne_results[:,1], c=cylinder_full[:,3])
        plt.title('TSNE: perp = %d' %(perp)+', iter = %d' %(iter), fontsize=sizeTitle)
        plt.show()
```

```python
#TSNE for a single set of hyper-parameters (full cylinder)
tsne = TSNE(n_components=2, verbose=1, perplexity=50, n_iter=1000)
tsne_results = tsne.fit_transform(cylinder_full[:,0:3])

fig = plt.figure(figsize = plt.figaspect(0.5))
matplotlib.rcParams.update({'font.size':sizeAxis})
ax = fig.add_subplot(1, 2, 1)
ax.scatter(tsne_results[:,0], tsne_results[:,1], c=cylinder_full[:,3])
plt.title('TSNE', fontsize=sizeTitle)

plt.show()
```

# #Twist dataset

```python
#Loading twist datasets
twist = np.load("twist.npy")
twistHole = np.load("twist_hole.npy")

fig = plt.figure(figsize = plt.figaspect(0.5))
matplotlib.rcParams.update({'font.size':sizeAxis})
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.scatter(twist[:,0], twist[:,1], twist[:,2], c=twist[:,3])
plt.title('Full', fontsize=sizeTitle)
ax = fig.add_subplot(1, 2, 2,  projection='3d')
ax.scatter(twistHole[:,0], twistHole[:,1], twistHole[:,2], c=twistHole[:,3])
plt.title('With Hole', fontsize=sizeTitle)
plt.show()
```

```python
#TSNE for various perplexity and iterations' number (twist)
for perp in range(5, 50, 5):
    for iter in range(250, 650, 50):
        tsne = TSNE(n_components=2, verbose=1, perplexity=perp, n_iter=iter)
        tsne_results = tsne.fit_transform(twist[:,0:3])
        #plots
        fig = plt.figure(figsize = plt.figaspect(0.5))
        matplotlib.rcParams.update({'font.size':sizeAxis})
        ax = fig.add_subplot(1, 2, 1)
        ax.scatter(tsne_results[:,0], tsne_results[:,1], c=twist[:,3])
        plt.title('TSNE: perp = %d' %(perp)+', iter = %d' %(iter), fontsize=sizeTitle)
        plt.show()
```

```python
#TSNE for a single set of hyper-parameters (twist)
tsne = TSNE(n_components=2, verbose=1, perplexity=50, n_iter=700)
tsne_results = tsne.fit_transform(twist[:,0:3])

fig = plt.figure(figsize = plt.figaspect(0.5))
matplotlib.rcParams.update({'font.size':sizeAxis})
ax = fig.add_subplot(1, 2, 1)
ax.scatter(tsne_results[:,0], tsne_results[:,1], c=twist[:,3])
plt.title('TSNE', fontsize=sizeTitle)

plt.show()
```
